```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
def read_csv_file(csv_path, csv_file_name):
    complete_path = os.path.join(csv_path, csv_file_name)
    try:
        df = pd.read_csv(complete_path)
        return df
    except FileNotFoundError:
        print(f'File not found at {complete_path}')
        return None
    except Exception as e:
        print(f'An error occurred while reading the CSV file: {e}')
        return None
```

```python
csv_path = 'csv/run/rrgcc/'

all_pmrp = 'all_pmrp.csv'

all_pmrp_run = read_csv_file(csv_path, all_pmrp)

if all_pmrp_run is not None:
    print('Data loaded successfully!')
```

```python
sns.scatterplot(x='distance_miles', y='moving_time_minutes', data=all_pmrp_run)
plt.xlabel('Distance (miles)')
plt.ylabel('Moving Time (minutes)')
plt.title('Scatter Plot: Distance vs. Moving Time')
plt.show()
```

```python
sns.histplot(all_pmrp_run['moving_time_minutes'], bins=20, kde=True)
plt.xlabel('Moving Time (minutes)')
plt.ylabel('Frequency')
plt.title('Histogram: Moving Time')
plt.show()
```

```python
sns.barplot(x='name', y='average_speed', data=all_pmrp_run)
plt.xlabel('Run Name')
plt.ylabel('Average Speed')
plt.title('Bar Plot: Average Speed for Each Run')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(x='name', y='total_elevation_gain', data=all_pmrp_run)
plt.xlabel('Run Name')
plt.ylabel('Total Elevation Gain')
plt.title('Box Plot: Total Elevation Gain for Each Run')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x='distance_miles', y='moving_time_minutes', hue='tavg', data=all_pmrp_run)
plt.xlabel('Distance (miles)')
plt.ylabel('Moving Time (minutes)')
plt.title('Scatter Plot: Distance vs. Moving Time with Temperature')
plt.legend(title='Temperature (tavg)')
plt.show()
```
