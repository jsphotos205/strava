##### PRCP - Precipitation
##### WT05 - Hail (may include small hail)
##### TOBS - Temperature at the time of observation
##### WT06 - Glaze or rime
##### SNWD - Snow depth
##### WT11 - High or damaging winds
##### WT01 - Fog, ice fog, or freezing fog (may include heavy fog)
##### TMAX - Maximum temperature

```python
rename_columns = {'WT01' : 'FOG',
                  'WT03' : 'THUNDER',
                  'WT04' : 'HAIL',
                  'WT06' : 'RIME',
                  'WT11' : 'WIND'}

weather_data =weather_data.rename(columns= rename_columns)

weather_data.head()
```

```python
weather_data.drop(['STATION', 'NAME', 'LATITUDE', 'LONGITUDE', 'ELEVATION'], axis=1, inplace=True)

weather_data = weather_data.fillna(0)
weather_data.head()
```

```python
weather_data['DATE'] = weather_data['DATE'].astype('datetime64[ns, UTC]')
```
