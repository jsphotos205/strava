```python
import pandas as pd
import folium
from ast import literal_eval
```

```python
df = pd.read_csv('csv/run/polyline.csv')
```

```python
df.head()
```

```python
second_run = df['decoded_polyline'][1]
second_run
```

```python
coordinates = literal_eval(second_run)
coordinates
```

```python
m = folium.Map(location=[coordinates[0][0], coordinates[0][1]], zoom_start=15)

folium.PolyLine(locations=coordinates, color='red').add_to(m)

m
```
