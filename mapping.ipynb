```python
import pandas as pd
import folium
from ast import literal_eval
```

```python
pmrp_run = pd.read_csv('csv/run/pmrp_run_data.csv')
sore_heel_loop = pd.read_csv('csv/run/sore_heel_data.csv')
lode_loop = pd.read_csv('csv/run/lode_loop_data.csv')
rrg_run = pd.read_csv('csv/run/rrg_run_data.csv')
```

```python
rrg_run.head()
```

```python
#access the start location from the dataframe
start_location = pmrp_run['start_latlng'][0]
print(start_location)
#remove square brackets
start_location = start_location[1:-1]
print(start_location)
#split the sting into lat and long values
lat, lng = start_location.split(', ')
print('Lat: ', lat, 'Lng:', lng)
#convert latitude and longitude to float
lat = float(lat)
lng = float(lng)

#add marker to the map
m = folium.Map(location=[lat,lng], zoom_start=15)
folium.Marker([lat,lng], popup='RRGCC Run Start Location').add_to(m)
```

```python
#access the start location from the dataframe
start_location = rrg_run['start_latlng'][0]
print(start_location)
#remove square brackets
start_location = start_location[1:-1]
print(start_location)
#split the sting into lat and long values
lat, lng = start_location.split(', ')
print('Lat: ', lat, 'Lng:', lng)
#convert latitude and longitude to float
lat = float(lat)
lng = float(lng)

r = folium.Map(location=[lat,lng], zoom_start=13)
folium.Marker([lat,lng], popup='Last RRG Run Location').add_to(r)
```

```python
for index, row in pmrp_run.iterrows():
    polyline= literal_eval(row['decoded_polyline'])

    feature_group = folium.FeatureGroup()

    folium.PolyLine(locations=polyline, color='red').add_to(feature_group)

    feature_group.add_to(m)
    
m.save('maps/pmrp_run.html')
m
```

```python
for index, row in sore_heel_loop.iterrows():
    polyline= literal_eval(row['decoded_polyline'])

    feature_group = folium.FeatureGroup()

    folium.PolyLine(locations=polyline, color='blue').add_to(feature_group)

    feature_group.add_to(m)
    
m.save('maps/sore_heel_loop.html')
m
```
