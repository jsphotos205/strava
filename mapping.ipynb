```python
import pandas as pd
import folium
from ast import literal_eval
```

```python
sore_heel_loop = pd.read_csv('csv/run/sore_heel_data.csv')
```

```python
sore_heel_loop.head()
```

```python
#access the start location from the dataframe
start_location = sore_heel_loop['start_latlng'][0]
print(start_location)
#remove square brackets
start_location = start_location[1:-1]
print(start_location)
#split the sting into lat and long values
lat, lng = start_location.split(', ')
print('Lat: ', lat, 'Lng:', lng)
#convert latitude and longitude to float
lat = float(lat)
lng = float(lng)

#add marker to the map
m = folium.Map(location=[lat,lng], zoom_start=13)
folium.Marker([lat,lng], popup='Start Location').add_to(m)
m
```
