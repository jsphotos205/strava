```python
from datetime import datetime
from meteostat import Stations, Daily
import pandas as pd
```

```python
run_data = pd.read_csv('csv/run/run_data.csv')
```

```python
start = datetime(2022, 1, 1)
end = datetime(2023, 7, 19)
```

```python
stations = Stations()
stations = stations.nearby(37.64471283182502, -83.71428340673447)
weather_station = stations.fetch(1)
```

```python
data = Daily(weather_station.iloc[:], start, end)
data = data.fetch()
```
